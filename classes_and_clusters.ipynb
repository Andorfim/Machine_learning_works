Генерация выборки с 4-мя кластерами:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize']=(12,8)
x,y = make_blobs(n_samples=100, random_state=1, centers = 4)
```

Визуализируем:

```python
plt.scatter(x[:,0], x[:,1])
```

Разбивка:

```python
kmeansModel = KMeans(n_clusters=4)
```

Обучим модель:

```python
kmeansModel.fit(x)
```

Получить метки кластеров:

```python
labels = kmeansModel.labels_
```

Визуализируем:

```python
plt.scatter(x[:,0], x[:,1], c=labels)
```

Кластеризация с использованием DBSCAN:

```python
clustering = DBSCAN(eps=1.3, min_samples=2).fit_predict(x)
plt.scatter(x[:,0], x[:,1], c=clustering)
```

Задание 2:


```python
data = pd.read_csv('Mall_Customers.csv')
data = pd.DataFrame(data)
x = data[['Spending Score (1-100)', 'Annual Income (k$)']]
x = x.to_numpy()
```

```python
criteries = []

for k in range(2,10):
  kmeansModel=KMeans(n_clusters=k, random_state=3)
  kmeansModel.fit(x)
  criteries.append(kmeansModel.inertia_)

plt.plot(range(2,10), criteries)
```

Метод K-средних:

```python
kmeansModel = KMeans(n_clusters=5)
kmeansModel.fit(x)
```

Визуализируем:

```python
plt.scatter(x[:,0], x[:,1], c=kmeansModel.labels_)
```

Метод DBSCAN:

```python
clustering = DBSCAN(eps=12, min_samples=10).fit_predict(x)
plt.scatter(x[:,0], x[:,1], c=clustering)
```
